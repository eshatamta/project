```python
nifty_ticker = '^NSEI'
start_date = dt.date(2019,1,1)
end_date = dt.date(2025,1,4)
```

```python
stock_data.xs(key = 'TCS.NS', axis =1, level= 1, drop_level = False)
```

```python
stock_data.xs(key = 'TCS.NS', axis =1, level= 1)
```

```python
tcs_stock = stock_data.xs(key= 'TCS.NS',axis= 1, level =1)
```

```python
plt.figure(figsize=(14,10))
sns.lineplot(data= tcs_stock[['High','Low']])
plt.title('stock metrics for TCS.NS')
plt.xlabel('Date')
plt.ylabel('Price')

plt.show()
```

```python
plt.figure(figsize=(14,10))
sns.lineplot(data= tcs_stock[['High','Low','Close','Open']])
plt.title('stock metrics for TCS.NS')
plt.xlabel('Date')
plt.ylabel('Price')

plt.show()
```

```python
tatasteel_stock = stock_data.xs(key= 'TATASTEEL.NS',axis= 1, level =1)
```

```python
plt.figure(figsize=(14,10))
sns.lineplot(data= tatasteel_stock[['High','Low','Close','Open']])
plt.title('stock metrics for TATASTEEL.NS')
plt.xlabel('Date')
plt.ylabel('Price')

plt.show()
```
